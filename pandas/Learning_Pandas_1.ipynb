### Loc and Iloc

🧠 Quick tip:

You want to...	Use this

Select by row/column position	=> iloc

Select by row/column label => loc

```python
cof.loc[[0,1,5]]
```

```python
cof.loc[5:8, ["Day","Coffee Type" ]]   #only the coffee type from rows 5 to 8
```

Loc => Values based on Index

```python
cof.loc[[0,1], "Coffee Type"]
```

Iloc => Values based on position

```python
cof.iloc[[0,1],1]
```

You want to... | Use | Example

Get a row by label (e.g., index = 5) | loc | df.loc[5]  \

Get a row by position (e.g., first row) | iloc | df.iloc[0] \

Get a value by label | loc | df.loc[5, 'Name'] \

Get a value by position | iloc | df.iloc[0, 1]

Get rows with label slice (inclusive) | loc | df.loc[2:5]

Get rows with index slice (exclusive) | iloc | df.iloc[2:5]

```python
cof.loc[2,"Units Sold"] = 10
```

```python
cof.loc[2:6,"Units Sold"] = 10
```

# Lambda Function

```python
bios['heightcat'] = bios['height_cm'].apply(lambda x: 'Short' if x < 165
                                            else ('Average' if x < 185
                                                  else 'Tall'))
```

```python
def newcat(row):
  if row['height_cm'] < 175 and row['weight_kg'] < 70:
    return 'Lightweight'
  elif row['height_cm'] < 205 and row['weight_kg'] < 110:
    return 'Normal'
  else:
    return 'Heavy'

bios['Category'] = bios.apply(newcat, axis = 1)
```

# Shift, Rank, Rolling and Cumsum functions

## Shift function

- shift(1): How was ur coffee revenue doing today than yesterday?

- shift(-1): Give each row the Revenue from tomorrow (the row below it)

```python
cof['yesterdayrev'] = cof['Revenue'].shift(1)
```

```python
cof
```

## Rank function

### What if we wanna rank the athletes based on their height?

```python
bios['hei'] = bios['height_cm'].rank()
```

```python
bios.sort_values(['hei'], ascending = False)
```

### Key differences in sort values

#### bios['hei'].sort_values(ascending=False)

- You're sorting just the 'hei' column, by itself.

- Returns a Series of the 'hei' values, sorted in descending order.

- Other columns are not touched or rearranged.

#### bios.sort_values(['hei'], ascending=False)

- You're sorting the entire DataFrame by 'hei', descending.

- The full rows move according to the 'hei' value.

- You get back a DataFrame, not just a single column.

## Cumsum function

- Used for running totals

### Revenue building up over time

```python
cof['total_revenue'] = cof['Revenue'].cumsum()
```

### Total per day (Monday's total resets when Tuesday starts).

```python
cof['daily_cumsum'] = cof.groupby('Day')['Revenue'].cumsum()
```

## Rolling function

#### Trying to find a combined latte cumsum for the past 3 days

- Rolling is used for rolling window.

```python
latte = cof[cof['Coffee Type'] == "Latte"].copy()

latte['dat2'] = latte['Units Sold'].rolling(3).sum()
```
