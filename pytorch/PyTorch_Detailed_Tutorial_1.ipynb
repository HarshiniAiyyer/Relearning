## Reshaping, stacking, Squeezing and unsqueezing tensors

- reshaping : fixing shapes / dimensions
- view -> return a view of the input tnesor but the same memory as the original tensor
- stacking -> multi tensors on top of the other (vstack) or side by side (hstack)

- squeeze -> removes all 1D from a tensor
- unsqueeze -> add 1D to a target tensor

- permute -> return a view of the input with dimensions permuted (swapped) in a certain way


````python
import torch

x = torch.arange(1.,10.)
x, x.shape
````

##### Reshape

````python
#add an extra dimension

xr = x.reshape(1,9)
xr, xr.shape, xr.ndim
````

##### View

````python
z = x.view(1,9)
z.shape
````

````python
z[:,0] = 5
z, x
````

##### Stacking

- hstack uses dim = 0,
- vstack uses dim = 1

````python
xs = torch.stack([x,x,x,x], dim = 1)
xs
````

##### Permute

- Used widely with images

````python
y = torch.rand(2,3,4)
y.size()
````

````python
torch.permute(y,(2,0,1)).size()
````

````python
xorig = torch.rand(size = (224, 224, 3))

xperm = xorig.permute(2,0,1)  #swaps the position of the axis

print(f"Previous shape: {xorig.shape}")
print(f"New shape: {xperm.shape}")
````

## Random Seed - Reproducibility in Pytorch

How a NN learns

- start with random nos -> then tensor operations -> update random nos to try and make them better representations of the data
-> and so on

- To reduce randommness in NN, and Pytorch, comes the concept of a ``random seed``

- Essentially what the random seed does is the 'flavor' the randomness. Or pseudo-randomness or generative randomness

````python
import torch

randtensA = torch.rand(3,4)
randtensB = torch.rand(3,4)

randtensA, randtensB
````

````python
print(randtensA == randtensB)
````

#### Random but reproducible tensors

````python
import torch

random_seed = 42  #we have different flavors - 42, 75 etc..

torch.manual_seed(random_seed)
randtensC = torch.rand(3,4)

#use manual seed everytime because it doesnt work for all together
torch.manual_seed(random_seed)
randtensD = torch.rand(3,4)

print(randtensC, randtensD)
print(randtensC == randtensD)
````
